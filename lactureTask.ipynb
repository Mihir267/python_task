```python
#INHERITANCE

class parent:
  def property(self):
    self.car = 'BMW'

class child(parent):
  def get_property(self):
    print(self.car)
```

```python
c.get_property()
```

```python
class parent:
  all_properties = dict()

  def property(self, **items):
    self.all_properties.update(items)

class child(parent):
  def get_property(self):
    return self.all_properties
```

```python
c = child()
```

```python
c.property(car = 'thar', bike = 'BMW', money = '10000')
```

```python
c.get_property()
```

```python
c.property(car = ['Thar','BMW','Audi'], bike = 'BMW', money = '10000')
```

```python
c.get_property()
```
